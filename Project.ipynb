```python
a = [1, 3, 4, 2, 7, 5, 8,6]
n = len(a)
def findMedian(a, n):
# First we sort the array
    sorted(a)
 
# check for even case
    if n % 2 != 0:
        return float(abs[int(n/2)])
 
    return float((a[int((n-1)/2)] +
                  a[int(n/2)])/2.0)
# Function call
print("Median =", findMedian(a, n))
```

```python
# 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less,
# it should be quoted as small. If the total bedrooms is 11 or more but less than 1000,
# it should be medium, otherwise it should be considered large.

#df1['total_bedroom_size'] =['small' if x <=10 else 'large' for x in df1['total_bedrooms']]
#df1.head()

import numpy as np
conditions = [
              (housing['total_bedrooms'] <=10),
              (housing['total_bedrooms'] >=11)&(housing['total_bedrooms'] <=1000),
              (housing['total_bedrooms'] > 1000)
             ]
bedroom_size = ['small','medium','large']

housing['total_bedroom_size']=np.select(conditions,bedroom_size)

housing
```
